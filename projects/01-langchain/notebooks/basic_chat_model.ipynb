# Chat Model

```python
from dotenv import load_dotenv
from pprint import pprint
from langchain.chat_models import init_chat_model

load_dotenv()
```

```python
model = init_chat_model(model='gpt-5-nano')

messages = [
    ("human", "Who is the best soccer player of all times?")
]
response = model.invoke(messages)
```

```python
pprint(response.content)
```

```python
# To have a continuous back and forth conversation you need to pass the entire history of the conversation to the model every time you ask a new question.
# Add the AI's response to your history list
messages.append(("ai", response.content))
```

```python
# Add a new follow-up question

messages.append(("human", "A best criteria is total goals"))
```

```python
# Invoke the model again

response2 = model.invoke(messages)
pprint(response2.content)
```
